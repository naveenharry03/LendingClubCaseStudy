# Data Cleaning and Manipulation

```python
# Extracting the column which has all rows as null values

data_with_all_nullvalues = data.isnull().sum()
data_with_full_column_nullvalues = data_with_all_nullvalues[data_with_all_nullvalues>=39717]
len(data_with_full_column_nullvalues)
```

```python
# Dropping those 54 columns from the dataset

data.drop(columns = data_with_full_column_nullvalues.index, inplace = True)
```

```python
# Again now checking the Rows and columns of the given dataset
    
data.shape
```

```python
# Checking the remaining null values in the dataset

data.isnull().sum()
```

```python
# desc - 12940 , mths_since_last_delinq - 25682 , mths_since_last_record - 36931 ,next_pymnt_d - 38577
# As these four columns has more no of null values when compared to the other columns.
# if we proceed with this columns means it will not add any value to the final outcome.
# So dropping these 4 columns\n",
    
Data_with_more_nullvalues = data[['desc','mths_since_last_record','mths_since_last_delinq','next_pymnt_d']]

data = data.drop(Data_with_more_nullvalues,axis = 1)
```

```python
# Again now checking the Rows and columns of the given dataset

data.shape
```

```python
# These 3 columns have unique values and also does not provide any valid informations to the analysis.
data_with_unnecessary_columns1 = data[['id','member_id','url']]

# This 2columns has more errors like same title is repeated with some spaces,special characters , dates etc. 
# so it is considered as new count when we use value_counts.
# Also it is alphanumeric and contains 2459 and 11 null values. 
data_with_unnecessary_columns2 = data[['emp_title','title']]
                                       
# These 9 Columns has single value for all the rows.(ex - In application type - all rows as value 'individual')
# These kinds of datas will not provide any proper analysis. 
data_with_unnecessary_columns3 = data[['pymnt_plan','policy_code','application_type','initial_list_status',
                                       'collections_12_mths_ex_med','acc_now_delinq','chargeoff_within_12_mths','delinq_amnt',
                                       'tax_liens']]    
                                       
# These 2 columns has more no of values to a single item when value_counts used
# For 'out_prncp' - it has 38577 values as 0.00
# For 'out_prncp_inv'  - it has 38577 values as 0.00
data_with_unnecessary_columns4 = data[['out_prncp','out_prncp_inv']]
```

```python
# Hence Removing all the Unnecessary columns.

data = data.drop(data_with_unnecessary_columns1,axis = 1)
data = data.drop(data_with_unnecessary_columns2,axis = 1)
data = data.drop(data_with_unnecessary_columns3,axis = 1)
data = data.drop(data_with_unnecessary_columns4,axis = 1)
```

```python
# Again now checking the Rows and columns of the given dataset
data.shape
```

```python
# At the time of approving the loan , the customer behaviour details will not be available. 
# So we need to remove those customer behaviour columns

data_with_customerbehaviour_details = data[['delinq_2yrs','earliest_cr_line','inq_last_6mths','open_acc','pub_rec',
                                           'revol_util','total_acc','revol_bal','total_pymnt','total_pymnt_inv',
                                            'total_rec_prncp','total_rec_int','total_rec_late_fee','recoveries',
                                           'collection_recovery_fee','last_pymnt_d','last_pymnt_amnt','last_credit_pull_d']]               
```

```python
# Hence Removing all the customer behaviour columns.

data = data.drop(data_with_customerbehaviour_details,axis = 1)
```

```python
# Since we are doing the case study to identify and provide the recommendations for the defaulters vs non defaulters.
# That is only for fully paid vs charged off
# So We will not be aware of the current loan status. So removing those rows from the dataset
# Drop those 1140 rows

data = data.loc[~(data['loan_status'] == 'Current')]
```

```python
# Checking the total counts of the current loan status is zero

len(data[data['loan_status'] == 'Current'])
```

#### Firstly taking public_rec_bankruptcies column to clear the null values. Even with or without imputation , this will not affect the analysis.

```python
# Checking the total counts of the pub_rec_bankruptcies which has null values

len(data[data['pub_rec_bankruptcies'].isnull()])

# Drop those 697 rows which has null values for the pub_rec_bankruptcies column

data = data.loc[~(data['pub_rec_bankruptcies'].isnull())]

# Checking the total counts of the emp_length which has null values has become zero

len(data[data['pub_rec_bankruptcies'].isnull()])
```

```python
# Moving next to the emp_length column. First checking the unique counts of each object
data['emp_length'].value_counts()
```

```python
# Removing the certain characters from the string to make it as a int format so that it will be used in quantitative analysis.
data['emp_length'] = data['emp_length'].str.replace(' year','').str.replace('s','').str.replace('< 1','0').str.replace('+','')

# Checking the column fields are properly replaced wit the above code

data['emp_length'].value_counts()
```

```python
# Checking the total counts of the emp_length which has null values

len(data[data['emp_length'].isnull()])

# Drop those 1033 rows which has null values for the emp length column

data = data.loc[~(data['emp_length'].isnull())]

# Checking the total counts of the emp_length which has null values has become zero

len(data[data['emp_length'].isnull()])
```

```python
# Checking the remaining null values in the dataset
data.isnull().sum()
```

## All null Values are cleared either by removing or replacing it with common values

### Moving to Standardisation of all columns

```python
data.info()
```

```python
# Now the next step is to change the datatype from object to int64 for emp_length. 
# By converting that we can compare this with the other numerical variables for our upcoming analysis

data.emp_length = data.emp_length.astype("int64")
data.emp_length.dtype
```

```python
# Now the next step is to change the datatype from object to int64 for term by removing certain characters in the string.

data['term'] = data['term'].str.rstrip(' months').astype('int')

data['term'].dtype
```

```python
# Now the next step is to change the datatype from object to float for int rate by removing (%) in the string.

data['int_rate'] = data['int_rate'].str.rstrip('%').astype('float')

data['int_rate'].dtype
```

```python
# Now the next step is to change the datatype from object to date for issue_d

data['issue_d'] = pd.to_datetime(data['issue_d'], format = '%b-%y')
```

```python
# Deriving new columns by grouping the distribution of quantitative variables to a certain range which helps to get more 
# insights from these. 

west = ['CA', 'OR', 'UT','WA', 'CO', 'NV', 'AK', 'MT', 'HI', 'WY', 'ID']
south_west = ['AZ', 'TX', 'NM', 'OK']
south_east = ['GA', 'NC', 'VA', 'FL', 'KY', 'SC', 'LA', 'AL', 'WV', 'DC', 'AR', 'DE', 'MS', 'TN' ]
mid_west = ['IL', 'MO', 'MN', 'OH', 'WI', 'KS', 'MI', 'SD', 'IA', 'NE', 'IN', 'ND']
north_east = ['CT', 'NY', 'PA', 'NJ', 'RI','MA', 'MD', 'VT', 'NH', 'ME']

def addr_state(state):
    if state in west:
        return 'West'
    elif state in south_west:
        return 'SouthWest'
    elif state in south_east:
        return 'SouthEast'
    elif state in mid_west:
        return 'MidWest'
    elif state in north_east:
        return 'NorthEast'
        
data['addr_state_grp'] = data['addr_state'].apply(addr_state)
```

```python
# To check the newly created columns are exist in the dataset

data['addr_state_grp'].value_counts()
```
